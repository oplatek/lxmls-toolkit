Consider the function $f (x) = (x + 2)^2 − 16 exp( 􏰂−(x − 2)^2)$. Make a function that computes the function value given x.

The derivative is 2 * x + 4 + exp(- (x - 2) ^2) * (32*x - 64)

```python
%matplotlib inline
import matplotlib.pyplot as plt

import numpy as np


def f(x):
    return (x+2) ** 2 - (16 * np.exp(-(x-2)**2))

x = np.linspace(-8,8,100)
y = f(x)
plt.plot(x, y)


def get_grad(x):
    return 2 * x + 4 + np.exp(- (x - 2) **2) * (32*x - 64)
    
    

def gradient_descent(start_x, func, grad): 
    # Precision of the solution
    prec = 0.0001
    #Use a fixed small step size step_size = 0.1
    step_size = 0.1
    #max iterations
    max_iter = 100
    x_new = start_x
    res = []
    for i in xrange(max_iter):
        x_old = x_new
        #Use beta egual to -1 for gradient descent
        x_new = x_old - step_size * grad(x_new) 
        f_x_new = func(x_new)
        f_x_old = func(x_old) 
        res.append([x_new,f_x_new]) 
        if(abs(f_x_new - f_x_old) < prec):
            print "change in function values too small, leaving"
            return np.array(res)
    print "exceeded maximum number of iterations, leaving" 
    return np.array(res)
```

```python
x_0 = -8
res = gradient_descent(x_0, f, get_grad)
plt.plot(x, y)
plt.plot(res[:,0], res[:,1], '+')
plt.show()
```

```python
x_0 = 8
res = gradient_descent(x_0, f, get_grad)
plt.plot(x, y)
plt.plot(res[:,0], res[:,1], '+')
plt.show()
```
